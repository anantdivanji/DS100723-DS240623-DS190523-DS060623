## Data Cleaning Ideas:

1. Deck column having lot of missing values. Action : Delete the deck the column {Done}
2. "sex" and "who" columns contain same information. Action : Delete any one of the column {Done}
3. PassengerID - Set it as an index {Done}
4. "survived" and "alive" contain same info. Action : Delete any one of the column. {Done}
5. "pclass" and "class" contain same info. Action : Delete any one of the column. {Done}
6. "embarked" and "embark_town" contain same info. Action : Delete any one of the column. {Done}
7. Round the Fare column to 2 decimal places {Done}
8. Capitalize the value of the column {Done}
9. Replace the values of survived 0: "No" and 1 : "Yes" {Done}
10. Alone column - Drop {Done}

```python
## 1. Delete the deck column from the data

data = data.drop("deck", axis = 1)
data.head()
```

```python
ls = ["who", "pclass", "embarked", "alive"]

data = data.drop(ls, axis = 1)
data.head()
```

```python
## Passenger ID
## If mean is not making sense then no other arithematic operation will make sense.
## Even value counts not giving any proper information. 
## 1. Either we can drop the passengerid column 2. Set the PassengerID column to Index
## Setting PassengerID column to index

data = data.set_index("PassengerID")
data.head()
```

```python
## Does having alone column in the data is useful or unuseful.
- Useful - 50%
- Unuseful - 50%
```

```python
## considering a case where alone is not useful. hence drop alone.

data = data.drop("alone", axis = 1)
data.head()
```

```python
data["fare"] = np.round(data["fare"], 2)
data.head()
```

```python
## Capitalize the values of sex column 

data['sex'] = data['sex'].str.capitalize()
data.head()
```

```python
## Replace the values of survived column

dic = {0 : "No", 1 : "Yes"}

data['survived'] = data['survived'].replace(dic)
data.head()
```

11. Capitalize all the column names
12.

```python
data.columns.str.capitalize()
data.head()
```

```python
## Capitalize all the column names 

data.columns = data.columns.str.capitalize()
data.head()
```

```python
## Rename the column adult_male and Embark_town
dic = {"Adult_male" : "AdultMale", "Embark_town" : "EmbarkTown"}

data = data.rename(dic, axis = 1)
data.head()
```

```python
## rename sibsp and parch

dic = {"Sibsp" : "Siblings/Spouse Count", "Parch" : "Parents/Children Count"}

data = data.rename(dic, axis = 1)
data.head()
```

## Save the clean data in your system
