# Feature Encoding :
- Converting the text columns(object columns) into numerical data is called Feature Encoding

```python
data = pd.read_csv("Data.csv")
data
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'most_frequent')
si
```

```python
data[['Country', 'Age', 'Salary']] = si.fit_transform(data[['Country', 'Age', 'Salary']])
data
```

```python
df = data.copy()
df
```

```python
X = df.drop("Purchased", axis = 1)
y = df['Purchased']
```

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc
```

```python
knc.fit(X, y)
```

- When you keep the text data as it is and apply ML algorithm on the data, it raises a value error.
- To get rid of this we perform Encoding(converting the values of the column to numerical)

### Types of Feature Encoding:
1. Manual Encoding
2. OneHotEncoding
3. LabelEncoding

# 1. Manual Encoding

```python
df
```

```python
dic = {"No" : 0, "Yes" : 1}
dic
```

```python
df["Purchased"] = df["Purchased"].replace(dic)
df
```

# One Hot Encoding:

- It applied on columns having nominal data. Nominal data the one which has no order in it.

```python
df
```

```python
df.info()
```

```python
df[['Age', 'Salary']] = df[['Age', 'Salary']].astype('float')
df.info()
```

```python
# get duumies function to perform one hot encoding

pd.get_dummies(df)
```

```python
df
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
dic = {"No" : 0, "Yes" : 1}
dic1 = {"Male" : 0, "Female" : 1}
dic2 = {"Lunch" : 0, "Dinner" : 1}
```

```python
tips['sex'] = tips['sex'].replace(dic1)
tips['smoker'] = tips['smoker'].replace(dic)
tips['time'] = tips['time'].replace(dic2)
```

```python
tips.head()
```

```python
tips['day'].unique()
```

```python
pd.get_dummies(tips['day'])
```

# Label Encoding on the day column

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le
```

```python
tips['day'] = le.fit_transform(tips['day'])
```

```python
tips.head()
```

```python
tips['day'].unique()
```

# One hot encoding using sklearn One Hot encoder function

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe
```

```python
df_ohe = ohe.fit_transform(df[['Country']])
df_ohe
```

```python
df_ohe.toarray()
```

```python
df_ohe = pd.DataFrame(df_ohe.toarray(), columns = ['France', 'Germany', 'Spain'])
df_ohe
```

```python
df
```

```python
df = pd.concat([df, df_ohe], axis = 1)
df
```

# Note:

- Apply One hot Encoding on Nominal data like country names, employee names, etc
- Apply Label Encoding on Ordinal data like days of the week, Grades of the students(A, B, C ,D ,E), designation of employees, etc
