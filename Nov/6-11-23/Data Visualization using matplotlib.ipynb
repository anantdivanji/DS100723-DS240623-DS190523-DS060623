# Data Visualization:

- Creating Charts, Visuals from the data to represent the insights in a more presentable manner.
- Matplotlib and Seaborn.
- Matplotlib and Seaborn is built on top of matplotlib.
- Matplotlib is more of technical library.
- Seaborn requires less code and also it provides more good looking charts compared to matplotlib.
- In matplotlib library, pyplot sub library provides the functions for creating visualization

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [20, 25, 40, 60]

## plot
plt.figure(figsize = (4, 3))  # control the size of the chart
plt.plot(x, y)    ## create the visualizaton. By default a line chart is created.
plt.show()    ##Get rid of the message displayed in the output
```

```python
## plot
plt.figure(figsize = (5, 3), facecolor = "y", edgecolor = "r", linewidth = 7)  
plt.plot(x, y, color = "green", linestyle = "-", marker = "o", linewidth = 2, markersize = 9)
plt.title("Line Chart displaying Covid Cases", size = 14, color = "red")
plt.xlabel("No. of Days", color = "blue", size = 12)
plt.ylabel("No. of Cases", color = "blue", size = 12)
plt.ylim(10, 120)
#plt.xlim(15, 35)
plt.xticks(x, labels = ["0-10", "10-20", "20-30", "30-40"])
plt.legend(["Cases by Days"])
plt.grid()
plt.show()

plt.savefig("Line Chart.png")
```

# Plotting using Matplotlib

1. plt.plot()
2. ax.plot()
- ax stands axes class which most basic and flexible unit for creating subplots. 

```python
x = [10, 20, 30, 40]
y = [20, 25, 40, 60]
```

```python
from matplotlib.figure import Figure

fig = plt.figure(figsize = (5, 4))

ax = fig.add_axes([1, 1, 1, 1])

ax1 = ax.plot(x, y)
ax2 = ax.plot(y, x)
plt.legend(["Male", "Female"], loc = 'best')
```

## Plot a pie chart

```python
holdings = [20, 55, 5, 17, 3]
company = ["Apple", "Google", "Tesla", "Alphabet", "Amazon"]

## Design
clrs_list = ['g', 'r', 'y', 'b', 'm']
exp_list = [0, 0.2, 0.3, 0, 0.2]
```

```python
plt.figure(figsize = (6, 6))
plt.pie(holdings, labels = company, colors = clrs_list, explode = exp_list)
plt.title("Stock Portfolio")
plt.show()
```

```python
values = [20, 20, 20, 20, 20]
clrs_list = ['g', 'r', 'y', 'b', 'm']
company = ["Apple", "Google", "Tesla", "Alphabet", "Amazon"]
exp_list = [0, 0.2, 0, 0.2, 0]
```

```python
plt.figure(figsize = (6, 6))
plt.pie(values, labels = company, colors = clrs_list, explode = exp_list)
plt.title("Stock Portfolio")
plt.show()
```

## Plot a Bar Chart

```python
Profit = [20, 35, 57, 89, 47]
Company = ["Tesla", "Amazon", "Google", "Alphabet", "Reliance"]
plt.bar(Company, Profit)
plt.title("Companies with their Profit")
plt.ylabel("Profit( Million $)")
```
