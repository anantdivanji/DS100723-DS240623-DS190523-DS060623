# 1. What is Machine Learning?

- In simple words, ML can be defined as teaching the machine to learn from the historical data.
- Its a study of teaching a computer program/an algorithm how to progressively improve upon a set that the is given to it.


## Traditional way of programming(period before 2012 or 2010):
- Software Developers write the code to build the applications.

## Machine Learning(period after 2010 or so) :
- Teach the machine to write the code in order to find the pattern and develop an application.
- You supply the data to a ML algorithm, ML algorithm itself will learn from the data and develop conclusions out of the data

# Supervised Learning:

- It is widely used when you have both input and output available inside your data

### Ways to apply Supervised ML on the dataset:

- We divide the data into 2 parts:
    1. Training data
    2. Test data
- How to divide into train and test data?
- The idea the training data should more and test data should be less(70% train and 30% test)
(80% train and 20% test)
- Training data is where the machine Learning algorithms are applied to learn the patterns from the data.
- Test data is used to check how good the ML algorithm was able to learn the patterns.

# Terminology of ML

# Terminology of Supervised ML:

- Input Variables - Independent Variables or Independent Features
- Ouput Variable - Dependent Variable, Target, output

- eg: In iris dataset, sepal_length, sepal_widht, petal length and petal_width are all Independent variables. species is the output or target
- eg: In tips datasets, tip variable is the target or the output and all other variable are independent

## Note: For any dataset, There must be only 1 column which we call as target. We cannot have multiple columns as target.
