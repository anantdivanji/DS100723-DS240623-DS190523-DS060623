## Display the top 10 Ranking Countries

```python
top_10 = data[data["Rank"] <= 10]    ### data[condition]
top_10
```

## Sorting the top 10 countries output

```python
top_10.sort_values("Rank")

## We do not specify to sort in ascending but still the output is sorted in ascending order. 
## It simply mean sort_values() will always sort in a ascending order
```

```python
## Sort the top 10 ranked countries in a descending order

top_10.sort_values("Rank", ascending = False)
```
