# Axes class

```python
x = [10, 25, 76, 89, 47, 34]
y = [15, 28, 35, 49, 53, 64]
z = ['A', 'B', 'C', 'D', 'E', 'F']
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2)
# plt.figure(figsize = (15, 9))

# plt.subplots_adjust(top=0.95)

axes[0, 0].plot(x, y, label = "subplot - 1")
axes[0, 1].scatter(x, y, label = "subplot - 2")
axes[1, 0].hist([2,2,3,3,3,3,3,4,4,5], label = 'subplot - 3')
axes[1, 1].bar(z,x, label = 'subplot - 4')

## Customize each plot
axes[0, 0].set_title('Line Plot', color = 'green')
axes[0, 1].set_title("Scatter Plot", color = 'red')
axes[1, 0].set_title("Histogram plot")
axes[1, 1].set_title("Bar plot")

plt.tight_layout()
plt.suptitle("Analysis of student exam scores")
plt.subplots_adjust(top=0.9)

plt.show()
```

# Numerical Univariate Analysis

```python
exam_scores = [75, 82, 90, 88, 78, 92, 85, 91, 79, 87, 88, 76, 83, 89, 84, 77, 92, 86, 80, 81]
exam_scores
```

# Numerical Data plots

```python
## Histogram plot : Spread of the data is given by histogram

plt.hist(exam_scores)
```

```python
import seaborn as sns

sns.histplot(exam_scores)
```

```python
## kdeplot plot : used to display the spread of the data

sns.kdeplot(exam_scores)
```

```python
## Boxplot : used to identify any outlier present in the data

sns.boxplot(exam_scores)
```

```python
## Violinplot

sns.violinplot(exam_scores, inner = 'sticks')
```

```python
## Swarm plot : 

sns.swarmplot(exam_scores)
```

```python
sns.ecdfplot(exam_scores)
```
