# Dataset from sklearn library

- Sklearn divides the inbuilt dataset into 4 categories
1. Toy Datasets - Work with tabular data(data is available in a table in rows and cols format)
2. Real - World Datasets - NLP and Computer Vision (Text and Images datasets)
3. Generate Datasets - You can use some functions like make_classification or make_regression to create your own datasets
4. Other Dataset - only 2 sample dataset available.

# Creating datasets with sklearn library

```python
## Create a regression dataset

from sklearn.datasets import make_regression
X, y = make_regression(n_samples = 20, n_features = 5)
```

```python
X
```

```python
y
```

```python
import pandas as pd
reg_data = pd.DataFrame(X)
reg_data
```

```python
pd.concat([reg_data, pd.DataFrame(y)], axis = 1)
```

- n_samples - used for defining the total rows to be present in the generated data
- n_features - used for defining the total columns to be present in X.
- Note: y will always have 1 column

```python
## Create a classification dataset

from sklearn.datasets import make_classification
X, y = make_classification(n_samples = 10, n_features = 5, n_classes = 2)
```
