# What is a Support Vector Machine (SVM)?
- SVM is a powerful ML algorithm used for linear and non linear classification & Regression & even for outlier detection.
- Applications of SVM include a variety of areas such as text classification, image classification, spam ham detection, handwriting identification, face detection, and anomaly detection.
- SVM's are highly adapetable and efficient in a variety of applications as they can manage high dimensional data and non linear relationships.
- SVM's are used for both classification and regression but they are very well suited for classification.

# Working of SVM:
- Main objective of SVM is to find an optimal hyperplane in an N - dimensional space that can seperate the data points in different classes in the
feature space.





- The term hyperplane will mean different for different types of data. Let's say you have only 2 variables, When you apply SVM on this kind of a data, the hyperplane will be a single line seperating the 2 variables
- For 3 variables, you apply SVM on this, the hyperplane will be a 2D plane that seperates the 2 classes.



- For more than 3 variables, the hyperplane will be harder to imagine. It can be a 3D plane , 4D plane or 5D plane etc

- The hyperplane tries that the margin between the closest points of different classes should be as maximum as possible.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
cancer = load_breast_cancer()
X = cancer.data[:, :2]    # Selecting all the rows of first 2 columns of the breast cancer data
y = cancer.target         # Select the target variable in y
```

```python
svm = SVC(kernel = "sigmoid")
svm.fit(X, y)
```

```python
# Plot the decision boundary (plot the hyperplane)

DecisionBoundaryDisplay.from_estimator(
    svm, 
    X,
    response_method = "predict",
    cmap = plt.cm.Spectral,
    alpha = 0.8,
    xlabel = cancer.feature_names[0],
    ylabel = cancer.feature_names[1],
)

plt.scatter(X[:, 0], X[:, 1], c = y, s = 20, edgecolors = "k")
plt.show()
```
