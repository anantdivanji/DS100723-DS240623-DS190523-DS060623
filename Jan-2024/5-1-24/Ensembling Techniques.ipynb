- Suppose i ask a question to a expert on a topic --> I will get a more convincing answer
- Suppose I ask the question to thousands of random people on a topic, then try to aggregate their answer.
- This we call it as wisdow of the crowd.

- Suppose Logistic Regression is best algorithm on a dataset. LR is giving me higher accuracy.
- Instead of just sticking to LR. I can train the data with a group of predictors(such as classifiers or regressors)
- You can often get better predictions than the individual couterpart.
- A group of predictors is called an ensemble, thus this technique is also called as Ensemble learning.
- An Ensemble Learning algorithm is called an Ensemble method

- For example, you train a group of Deicision Tree classifiers, each on a different random subset of the training set.
- to make predictions, you just obtain the predictions of all individual trees, then predict the class that the most votes.
- This ensemble of Decision Trees is called as Random Forest.
- Random Forest is simple but most powerful ML algorithms available today.

```python
# Seperate X and y
X = data.drop('target', axis = 1)
y = data['target']
```

```python
X = pd.get_dummies(X).astype('int')
X.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
X[["day", "duration", "age", "accountBalance"]] = mms.fit_transform(X[["day", "duration", "age", "accountBalance"]])
X.head()
```

```python
# Split the data into train and test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)   
```

```python
# apply Logistic Regression on the train set

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

## Observation:
- Individually when we are applying ML algorithms to the data we notice that accuracy changes.
- Above we have checked with only LogReg and DT, but we can also check with SVM, KNClassifier, etc...

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 300, max_leaf_nodes = 16)
rfc.fit(X_train, y_train)
```

```python
y_pred_frc = rfc.predict(X_test)
```

```python
accuracy_score(y_test, y_pred_frc)
```
